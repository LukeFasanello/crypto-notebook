```python
import requests
import pandas as pd
import matplotlib.pyplot as plt


def fetch_symbols():
    
    url = 'https://api.binance.us/api/v3/exchangeInfo'
    resp = requests.get(url)
    resp.raise_for_status()
    all_symbols = [s['symbol'] for s in resp.json()['symbols']]
    df = pd.DataFrame(all_symbols, columns = ['Symbol'])
    print(df.to_string())
    

fetch_symbols()
    
```

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt

def fetch_trades(symbol='BTCUSDT', limit=1000):
    """
    Fetch the last `limit` trades for `symbol` from Binance and
    return a DataFrame indexed by datetime, with float price & qty.
    """
    url = 'https://api.binance.us/api/v3/trades'
    resp = requests.get(url, params={'symbol': symbol, 'limit': limit})
    resp.raise_for_status()
    data = resp.json()

    # Build DataFrame
    df = pd.DataFrame(data)
    df['price'] = df['price'].astype(float)
    df['qty']   = df['qty'].astype(float)
    df['quoteQty']   = df['quoteQty'].astype(float)
    # Convert ms-since-epoch to pandas.Timestamp and set as index
    df['datetime'] = pd.to_datetime(df['time'], unit='ms')
    df.set_index('datetime', inplace=True)
    
    return df

# Fetch & inspect
df = fetch_trades()
df.head()


    





```
